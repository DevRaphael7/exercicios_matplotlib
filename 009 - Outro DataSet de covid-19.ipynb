```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"Tabelas\country_vaccinations_by_manufacturer.csv")
df.head()
```

# As vacinas

```python
df['vaccine'].unique()
```

# Os fabricantes de vacinas que mais aplicaram

```python
plt.bar(df['vaccine'], df['total_vaccinations'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x='vaccine', y='total_vaccinations', data=df)
plt.xticks(rotation=90)
plt.show()
```

# Variável location

```python
df['location'].unique()
```

# Número de vacinados nos United States

```python
United_States = df[['location', 'total_vaccinations']].query('location == "United States"')
United_States['total_vaccinations'].sum()
```

# Número de vacinados na Romania

```python
Romania = df[['location', 'total_vaccinations']].query('location == "Romania"')
Romania['total_vaccinations'].sum()
```

### os fabricantes de vacinas que mais foram aplicados na Romania

```python
Romania = df[['location', 'vaccine', 'total_vaccinations']].query('location == "Romania"')
sns.barplot(x='vaccine',y='total_vaccinations',data=Romania)
plt.xticks(rotation=60)
```

### Número de vacinados por data na Romania

```python
Romania = df[['location', 'date', 'total_vaccinations']].query('location == "Romania"')
plt.figure(figsize=(30, 160))
sns.barplot(x='total_vaccinations', y='date', data = Romania, orient='h')
plt.show()
```
