# Os jogos mais vendidos por ano (variável Global_Sales)

```python
tabela[['Year', 'Global_Sales']].head()
```

```python
pd.unique(tabela['Year'])
```

```python
x = pd.DataFrame(tabela['Year'].value_counts().keys(), columns=['Ano'])
x['Vendas'] = tabela['Year'].value_counts().values
ordenado = x.sort_values(by=['Ano'], ascending=False)
ordenado.head()
```

Para plotar o gráfico de jogos mais vendidos por ano, é necessário programação.

Algoritmo:
    
1 <-- Ordenar a variável 'Year'

2 <-- Criar duas listas para armazenar as observações das variáveis

3 <-- Fazer um laço de repetição, dentro do laço de repetição põe uma estrutura condicional onde fará o somatório das vendas de cada ano, por meio da coluna(variável) ordenada.

```python
lista_anos = []
lista_vendas = []

for i in ordenado['Ano']:
    y = tabela[['Year', 'Global_Sales']].query("Year == %f" % i)
    y = np.sum(y['Global_Sales'])
    lista_anos.append(i)
    lista_vendas.append(y)

vendidos_por_ano = pd.DataFrame(lista_anos, columns=['Ano'])
vendidos_por_ano['Vendas'] = lista_vendas
vendidos_por_ano
```

```python
plt.bar(vendidos_por_ano['Ano'], vendidos_por_ano['Vendas'])
plt.title("Vendas de jogos por ano\n", fontsize=20)
plt.xlabel("\nAnos", fontsize=12)
plt.ylabel("Número de vendas\n", fontsize=12)
plt.show()
```

O ano de 2006 a 2010 foram anos que tiveram maiores vendas de jogos. Sendo que 2008 foi o ano que mais vendeu.

```python
ordenado['Vendas'].max()
```

```python
vendidos_por_ano.sort_values(by='Vendas', ascending=False)
```

# Quais foram as desenvolvedoras que mais publicaram jogos?

```python
desenvolvedoras_jogos = pd.DataFrame(tabela['Publisher'].value_counts().keys(), columns=['Desenvolvedoras'])
desenvolvedoras_jogos['quant_Publisher'] = tabela['Publisher'].value_counts().values
desenvolvedoras_jogos.head(20)
```
