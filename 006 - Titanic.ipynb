# Analisando DataSet de um Titanic

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_dataSet = pd.read_csv(r"Tabelas\train.csv")
titanic_dataSet.head()
```

# Quantos sobreviveram? 

```python
titanic_dataSet[['Survived']].value_counts()
```

Um número grande de passageiros morreram, 549, e um número menor de sobreviventes comparado ao número de passageiros que morreram, sobreviveram.

```python
titanic_dataSet[['Survived']].value_counts().plot(kind='bar')
plt.xlabel("\nPassageiros (0) mortos e (1) vivos")
plt.ylabel("Número de mortes\n")
plt.show()
```

# Analisando a variável Sex do DataSet

```python
titanic_dataSet['Sex'].value_counts()
```

A maioria dos passageiros era homem, mais que o sexo feminino, 314.

```python
titanic_dataSet['Sex'].value_counts().plot(kind='barh')
plt.show()
```

# Analisando a variável Age do DataSet

```python
titanic_dataSet['Age'].value_counts().sort_values(ascending=False)
```

```python
plt.figure(figsize=(22, 18))
titanic_dataSet['Age'].value_counts().sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=60)
plt.xlabel("\nIdades")
plt.show()
```

Os passageiros eram na maioria jovens de 20 a 30 anos. Mas também tem bastante passageiros mais velhos.

```python
titanic_dataSet['Age'].min() , titanic_dataSet['Age'].max()
```

O mais velho da embarcação tinha 80 anos. o 0 talvez queira dizer que sejam recém-nascidos.

# Entre os sobreviventes, há mais homens ou mulheres?

```python
titanic_dataSet[['Survived', 'Sex']].query('Survived == 1 and Sex == "male"').count()
```

```python
titanic_dataSet[['Survived', 'Sex']].query('Survived == 1 and Sex == "female"').count()
```

Entre os sobreviventes a maioria é do sexo feminino

```python
titanic_dataSet[['Survived', 'Sex']].query('Survived == 1').value_counts().plot(kind='barh')
plt.xlabel("\nNúmero de sobreviventes")
plt.ylabel("Sexo dos sobreviventes\n")
plt.show()
```

# Idade dos sobreviventes

```python
plt.figure(figsize=(22, 18))
titanic_dataSet[['Survived', 'Age']].query('Survived == 1').value_counts().plot(kind='bar')
plt.xlabel("\nIdade dos sobreviventes")
plt.ylabel("Número de sobreviventes")
```

Idade dos sobreviventes mulheres

```python
titanic_dataSet[['Survived', 'Sex', 'Age']].query("Survived == 1 and Sex == 'female'").value_counts()
```

```python
titanic_dataSet[['Survived', 'Sex', 'Age']].query("Survived == 1 and Sex == 'male'").value_counts()
```

### Gráfico de dispersão nas idades dos sobreviventes

```python
titanic_dataSet.query("Survived == 1").plot(kind="scatter", x="Survived", y="Age")
plt.grid(True)
plt.show()
```

Houve um sobrevivente com 80 anos, o mais velho entre os sobreviventes.
