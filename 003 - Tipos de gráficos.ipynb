```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dicionario = {
    'Alunos': ['Raphael', 'Miriam', 'Matheus', 'Pedro'],
    'Notas': [7.5, 8.0, 9.0, 6.5]
    
}

notasAlunos = pd.DataFrame(dicionario)
notasAlunos
```

# Gráfico de barras vertical

```python
plt.bar(notasAlunos['Alunos'], notasAlunos['Notas'])
plt.show()
```

# Gráfico de barras horizontal

```python
plt.barh(notasAlunos['Alunos'], notasAlunos['Notas'])
plt.show()
```

# Gráficos de dispersão

```python
plt.scatter(notasAlunos['Alunos'], notasAlunos['Notas'])
plt.show()
```

# Gráfico de linhas

```python
plt.plot(notasAlunos['Alunos'], notasAlunos['Notas'])
plt.show()
```

# Boxplot ou diagrama de caixa

```python
notasAlunos.boxplot(column=['Notas'])
plt.show()
```

```python
plt.scatter(notasAlunos['Alunos'], notasAlunos['Notas'])
notasAlunos.boxplot(sym='bo')
plt.show()
```

```python
dicionario = {
    '1° Semestre':[4.0, 6.3, 5.5, 9.0, 7.5],
    '2° Semestre':[9.0, 7.0, 7.0, 6.0, 7.0]
}

notasAlunos = pd.DataFrame(dicionario)

notasAlunos.boxplot(sym='b+')
plt.show()
```

```python

notasAlunos['Média'] = (notasAlunos['1° Semestre'] + notasAlunos['2° Semestre'])/2

notasAlunos['Média'].plot(kind='hist')
plt.show()
```
